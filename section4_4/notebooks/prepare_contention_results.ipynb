```python
import numpy as np
from pprint import pprint
import matplotlib.pyplot as plt
```

```python
fname = '../data/Contention_results_with_coarse_grained_measurements.txt'
data = np.loadtxt(fname, skiprows=1)
print(data.shape)
```

```python
x = data[:,0]
y = data[:,1]
plt.plot(x,y, color='blue')
plt.grid(color='gray', linestyle='-', linewidth=.4)
plt.title('Process I/O Contention')
plt.xlabel('Contending Processes')
plt.ylabel('Read Latency [secs]')
# plt.savefig('../data/process_contention.png')
plt.show()
```
